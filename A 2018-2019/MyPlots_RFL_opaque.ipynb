## Results of RLF method

### Opaque Surface

Here I would like to present the procedure and results of my script that calulates the load of opaque surfaces using **RLF** *method*, you can find the details in [Polimi](www.polimi.it)

Now that I have calculated the loads, let me show them to you through a bar chart

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python


items1=[1,3,5]
items2=[2,4,6]
Labels = ["wall","floor","door"]
heatingLoads = [2581,3500,300]
coolingLoads = [2300,3400,250]
fig1 = plt.figure()
plt.bar(items1,heatingLoads,label="heating Load", color="r")
plt.bar(items2,coolingLoads,label="cooling Load")
plt.legend()

plt.xticks(items1,Labels)
```
